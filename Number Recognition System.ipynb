# Number Recognition System

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten
from keras.utils import to_categorical
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# Normalize pixel values to be between 0 and 1
X_train = X_train / 255.0
X_test = X_test / 255.0
```

```python
# Flatten the images
X_train_flat = X_train.reshape((X_train.shape[0], 28, 28, 1))
X_test_flat = X_test.reshape((X_test.shape[0], 28, 28, 1))

# One-hot encode the labels
y_train_one_hot = to_categorical(y_train)
y_test_one_hot = to_categorical(y_test)
```

```python
# Build the Neural Network
model = Sequential()
model.add(Flatten(input_shape=(28, 28, 1)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train_flat, y_train_one_hot, epochs=10, batch_size=32, validation_split=0.2)

```

```python
loss, accuracy = model.evaluate(X_test_flat, y_test_one_hot)
print(f'Test Accuracy: {accuracy * 100:.2f}%')
```

```python
# Choose a random test image
index = np.random.randint(0, len(X_test))
test_image = X_test_flat[index].reshape(1, 28, 28, 1)

# Predict the digit
prediction = model.predict(test_image)
predicted_label = np.argmax(prediction)

plt.imshow(X_test[index], cmap='gray')
plt.title(f'Actual: {y_test[index]}, Predicted: {predicted_label}')
plt.show()
```
